```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("data/synthetic-1.csv").values
plt.scatter(data[:, 0], data[:, 1], 2)
```

### Stwórzmy kod do rysowania wykresu elbow

```python
k_vec = []
init_vec =[]

for k in range(1,10):
    km = KMeans(n_clusters=k,n_init=1,init='k-means++').fit(data)
    k_vec.append(k)
    init_vec.append(km.inertia_)
```

```python
k_vec
```

```python
init_vec
```

```python
plt.plot(k_vec,init_vec)
```
