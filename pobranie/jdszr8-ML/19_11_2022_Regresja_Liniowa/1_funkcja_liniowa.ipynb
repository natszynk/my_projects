# Funkcja liniowa

\begin{equation*}
y=ax+b
\end{equation*}  
<I>a</I> - slope  
<I>b</I> - intercept  
<I>x</I> - input  
<I>y</I> - output  

## Zadanie
### napisz funkcję, która wyliczać będzie odpowiedź funkcji liniowej na podstawie wejścia oraz dwóch współczynników.

```python
def linear_function(a,b,x):
    return a*x+b
```

```python
a = 2
b = 1
x = 3
y = linear_function(a,b,x)
```

```python
print(f'y = {a}*{x}+{b} = {y}')
```

### Narysuj wykres funkcji liniowej na podstawie danych syntetycznych.

```python
import numpy as np
```

```python
x = np.arange(-10, 10, 1)
x
```

```python
y = linear_function(a,b,x)
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

### Stwórz wykres nakładający na siebie kilka funkcji liniowych o różnych współczynnikach slope

```python
#definiujemy różne wartości współczynnika slope
a = [-2, -1, 0, 1, 2]
b = 1
```

```python
def draw_lin_func_slope(a,b,x):
    y = linear_function(a,b,x)
    plt.plot(x,y,label=str(a))
```

```python
for ai in a:
    draw_lin_func_slope(ai,b,x)
plt.legend(loc='lower center')
plt.title("funkcja liniowa z różnymi wartościami współczynnika slope")
plt.show()
```

### Stwórz wykres nakładający na siebie kilka funkcji liniowych o różnych współczynnikach intercept

```python
#definiujemy różne wartości współczynnika intercept
a = 1
b = [-2, -1, 0, 1, 2]
```

```python
def draw_lin_func_intercept(a,b,x):
    y = linear_function(a,b,x)
    plt.plot(x,y,label=str(b))
```

```python
for bi in b:
    draw_lin_func_intercept(a,bi,x)
plt.legend(loc='lower center')
plt.title("funkcja liniowa z różnymi wartościami współczynnika intercept")
plt.show()
```

## Odległości między dwiema funkcjami liniowymi o różnych współczynnikach slope
tip: matplotlib.pyplot.errorbar()

```python
x = np.arange(-7,7)
```

```python
# Dwa różne wykresy f-cji liniowej, z dwoma różnymi a
a1 = 1
a2 = 2
b=2
y1 = linear_function(a1,b,x)
y2 = linear_function(a2,b,x)
yerr = y2 - y1 # yerr = y-error
```

```python
y1
```

```python
y2
```

```python
yerr
```

```python
# Czy większe od 0, czy większe niż n
lolims = np.greater(yerr,0)
lolims
```

```python
# Czy niższe od 0, czy niższe od n
uplims = np.less(yerr,0)
uplims
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(x,y1,marker ='o', markersize=6, label='y1')
plt.plot(x,y2,marker ='o', markersize=6, label='y2')
plt.errorbar(x,y1,yerr=np.abs(yerr),linestyle=None,lolims=lolims, uplims=uplims)
plt.legend()
plt.grid()
plt.show() # lub pass (nie wyświetla informacji o obiekcie)
```
